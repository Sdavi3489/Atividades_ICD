```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sns

def grafico_notas():
  #Carregando o arquivo csv por um link do Github
  notas = pd.read_csv('https://raw.githubusercontent.com/Sdavi3489/Atividades_ICD/main/ratings.csv' ,encoding='cp1252')
  
  #Colunas traduzidas e renomeadas para português
  notas.columns = ['usuarioID','filmeID','nota','momento']

  #Média de todos os filmes
  media_filme1 = notas.query('filmeID==122920').nota.mean() #Média do filme Captain America - Civil War (2016)
  media_filme2 = notas.query('filmeID==175485').nota.mean() #Média do filme Death Note (2017)
  media_filme3 = notas.query('filmeID==188301').nota.mean() #Média do filme Ant-Man and the Wasp (2018)
  media_filme4 = notas.query('filmeID==79879').nota.mean()  #Média do filme Piranha (Piranha 3D)
  media_filme5 = notas.query('filmeID==122918').nota.mean() #Média do filme Guardians of the Galaxy 2 (2017)
  media_filme6 = notas.query('filmeID==175475').nota.mean() #Média do filme The Emoji Movie (2017)

  #Dataframe contendo todos os filmes e média dos filmes aprovados e reprovados
  df_filmes = {
      'Filmes' : ['Captain America - Civil War (2016)','Death Note (2017)','Ant-Man and the Wasp (2018)','Piranha (Piranha 3D)','Guardians of the Galaxy 2 (2017)','The Emoji Movie (2017)'],
      'Média_aprov': [media_filme1,0,media_filme3,0,media_filme5,0], #Média de cada filme aprovado
      'Média_rep' : [0,media_filme2,0,media_filme4,0,media_filme6] #Média de cada filme reprovado
  }
  pd.DataFrame(df_filmes)

  #Definindo a proporção do gráfico
  plt.figure(figsize=(8,5))

  #Definido título para o gráfico
  plt.title('Média das avaliações dos filmes',fontsize='20')

  #Definindo os labels do eixo x e eixo y
  plt.xlabel('Título',fontsize='15')
  plt.ylabel('Média da nota dos filmes',fontsize='15')
  
  #Plotando o gráfico de barras
  plt.bar(data = df_filmes, x = df_filmes['Filmes'], height = df_filmes['Média_aprov'],color=['g'],label='Gostei')
  plt.bar(data = df_filmes, x = df_filmes['Filmes'], height = df_filmes['Média_rep'],color=['r'],label='Não gostei')

  #Definindo labels do eixo x e colocando na vertical
  plt.xticks(df_filmes['Filmes'], rotation='vertical')

  #Adicionando legenda ao lado direito do gráfico
  plt.legend(ncol=1, loc='upper right',bbox_to_anchor=(1.240,1))

  #Salva o gráfico em alta resolução
  plt.savefig('media_filmes.png', dpi=300, bbox_inches='tight')
  return plt.show()
```

```python
grafico_notas()
```
